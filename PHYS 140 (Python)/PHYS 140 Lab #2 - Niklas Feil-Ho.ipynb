# Chapter 2 Data Analysis: Thermodynamics of Electrons on a Lattice

```python
#b
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('Hubbard.dat')
temp = data[0:50,1] #c
print (temp)


```

```python
#d
chem = data[0:42549:50, 0]
print (chem)
```

```python
#e
density = data[:, 2]
energy = data[:, 3]
E = energy.reshape(851,50)
print (E)
```

```python
n = density.reshape(851, 50)
print (n)
```

```python
#f
y = E[425, :]
x = temp
plt.xlabel("Temperature")
plt.ylabel("Energy")
plt.plot(x,y)
plt.show()
```

```python
y = n[425, :]
x = temp 
plt.xlabel("Temperature")
plt.ylabel("Density")
plt.plot(x,y)
plt.show()
```

```python
#g
def get_mu_ix(rho):
    
    
    return np.argmin(abs(n - rho),axis = 0)

get_mu_ix(.9)
```

```python
#h
rhos = np.linspace(0.65,0.95,7)

for rhoval in rhos: 
    mu_idx = get_mu_ix(rhoval)
    eRho = E[mu_idx, :]
    
plt.xlabel("Temperature")
plt.ylabel("Energy")
plt.plot(temp, eRho)
plt.show()
```

```python
#i
rho_value = 0.8
mu_idx_rho = get_mu_ix(rho_value)
toFile = np.column_stack((temp, E[mu_idx_rho, :]))
np.savetxt('Temperature&Energy', toFile)
#cannot upload the txt because canvas only accepts ipynb
```
