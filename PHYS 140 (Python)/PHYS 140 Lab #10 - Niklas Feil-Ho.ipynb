# 1.1 Equilibrium positions of Molecule

```python
#b
import numpy as np
import matplotlib.pyplot as plt

def solve_positions(k1, k2, F):
    # Coefficient matrix for the system
    A = np.array([
        [k2 + k1, -k1, 0],
        [-k1, 2 * k1, -k1],
        [0, -k1, k1 + k2]
    ])
    # Right-hand side (forces) applying a small force to ensure non-trivial solution
    b = np.array([F, 0, -F])
    
    # Solve for the positions
    positions = np.linalg.solve(A, b)
    return positions

def calculate_length(positions):
    return np.abs(positions[-1] - positions[0])

# Ratio range for k1/k2
ratios = np.linspace(0.1, 10, 100)
lengths = []

k2 = 1  # Fixed k2 value
F = 0.1  # Small external force

for ratio in ratios:
    k1 = ratio * k2
    positions = solve_positions(k1, k2, F)
    total_length = calculate_length(positions)
    lengths.append(total_length)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(ratios, lengths, label='Total Length of the System')
plt.xlabel('k1/k2')
plt.ylabel('Total Length')
plt.title('Total Length of the System as a Function of k1/k2')
plt.legend()
plt.grid(True)
plt.show()
```
