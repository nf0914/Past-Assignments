# 1.3 Catalan Numbers

```python
def catalanNumbers(limit):
    catalanNums = [1]

    n = 0
    while True:
        nextCatalan = ((4 * n + 2) / (n + 2)) * catalanNums[-1]
        
        if nextCatalan <= limit:
            catalanNums.append(nextCatalan) #creating a list
            n += 1
        else:
            break
        
        
    return catalanNums

limit = 10**9 #1 billion limit
        
catalanNums = catalanNumbers(limit)
print("All Catalan Numbers less than or equal to 1 billion are:")
for numbers in catalanNums[1:]: #printing each new value on a separate line; making sure 1 is not printed twice
    print(numbers)
        
    

```

# 1.4 Fourier Synthesis of a Square Wave 

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 1000)

plt.figure(figsize=(10,6))
plt.plot(x,0.5 * (1 + np.sign(np.sin(x))), '-o', label = 'Sqaure Wave')

maxTerm = 10
constant = 0.5

for n in range(1, maxTerm + 1, 2):
    constant += (2 / (n * np.pi)) * np.sin(n * x)
    
plt.plot(x, constant, '-0', label = 'Fourier Series Approximation')

plt.xlabel('x')
plt.ylabel('A(x)')
plt.title('Square Wave and Fourier Series Approximation')
plt.legend()
plt.show()
```
