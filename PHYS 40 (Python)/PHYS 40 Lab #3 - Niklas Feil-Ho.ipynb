# 3.2 Plot a Gaussian 

```python
def func(x,sigma):
    g = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-x**2/sigma)
    return g
x = np.linspace(-5,5,100)
y1 = func(x,1.0)
y2 = func(x,1.5)
y3 = func(x, 2.0)
plt.plot(x,y1,'b',label='Sigma at 1')
plt.plot(x,y2,'k',label='Sigma at 1.5')  
plt.plot(x,y3,'r',label='Sigma at 2.0')
plt.axvline(x=1,ymin=0,ymax=1,color='b')
plt.axvline(x=1.5,ymin=0,ymax=1,color='k')
plt.axvline(x=2.0,ymin=0,ymax=1,color='r')
plt.xlabel('x') 
plt.ylabel('y') 
plt.title("Gaussian Function") 
plt.legend(loc='upper left')
import matplotlib
matplotlib.rcParams.update({'font.size': 2})
```
