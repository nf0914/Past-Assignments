# 10.1 Fourier Series and Transform 

```python
import numpy as np 
import matplotlib.pyplot as plt 

def square(x,T): 
    if x < T/2:
        return 1 
    elif x<=T:
        return -1
    elif (x > T): 
        return square(x-T,T) 
    elif (x < 0): 
        return square(x+T,T) 
```

```python
def fourier(x,N,T):
    sum=0
    v=1
    for k in range(1,N+1): 
        sum += 1/(2*k-1)*np.sin(2*np.pi*(2*k-1)*v*x)
    sum=(4/np.pi)*sum
    return sum   
```

```python
x =np.linspace(0,2,2048) 
y=np.zeros(2048) 
for k in range(len(x)): 
    y[k]=square(x[k],1) 
```

```python
y1 = fourier(x,3,2.0) 
y2 = fourier(x,9,2.0) 
y3 = fourier(x,18,2.0) 
plt.plot(x,y,'b',label='square') 
plt.plot(x,y1,'r',label='3 terms') 
plt.plot(x,y2,'g',label='9 terms') 
plt.plot(x,y3,'m',label='18 terms') 
plt.xlabel('x') 
plt.ylabel(r'f(x),f$_{fourier}$(x)') 
plt.grid() 
plt.legend()
```
