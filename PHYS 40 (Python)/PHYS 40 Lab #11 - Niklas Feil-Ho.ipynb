# 11.2 Random Walk 

```python
import numpy as np
# xi's walk is random becasue he has an equal chance of winning and losing
i=0
x_0=10
F=100
p=0.49

while x_0<F:
    path = np.random.uniform(low=0, high=1)
    if path <= p:
        x_0+=1
    elif path>= p:
        x_0-=1
    if x_0<=0:
        x_0+=10
    i+=1
    print(x_0,i)
    
    
```
