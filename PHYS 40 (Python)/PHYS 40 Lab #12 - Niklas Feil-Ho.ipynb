# 12.1 Fit Cosmic background 

```python
import numpy as np 
import matplotlib.pyplot as plt 
from scipy.optimize import curve_fit 


file=np.loadtxt('cmb-data-2.txt')
v= file[:,0] #frequency
intensity=file[:,1]



```

```python
c= 2.998e10
h= 6.6261e-27
k=1.3807e-16

def func(v,t):
    expression = (2*h*(v**3)*(c**2))/(np.exp((h*c*v)/(k*t))-1)
    return expression
```

```python
curve_fit = curve_fit(func,v,intensity,sigma=(1.0e-6)*np.ones(v.size),absolute_sigma=True)
```

```python
v
```

```python
intensity
```

```python
curve_fit
```

```python
fit=func(v,curve_fit[0])
plt.plot(v,intensity,'.')
plt.plot(v,fit)
plt.xlabel('Frequency') 
plt.ylabel('Intensity') 
plt.title("Microwave Radiation Intensity vs. Radiation")

Tfit=float(curve_fit[0])
Tfit_1=float(curve_fit[1])
output='T ='+'{:1.3f}'.format(Tfit)+' Uncertainty ='+'{:1.7f}'.format(Tfit_1)
print(output)
```

# 12.2 Sierpinski Triangle (Gasket) 

```python
import random as rn
points=np.zeros((3,2))
points[0,:]=(0,0)
points[1,:]=(1,0)
points[2,:]=(1/2,np.sqrt(3)/2)
```

```python
points
```

```python
def func():
    rando_1= rn.uniform(0,1)
    rando_2=rn.uniform(0,1)
    while not (rando_2 <= np.tan(np.pi/3)*rando_1 and rando_2<=-np.tan(np.pi/3)*rando_1+np.tan(np.pi/3)):
        return np.array([rando_1,rando_2])
```

```python
func()
```

```python
N=10000
r=np.zeros((2,N))
r[:,0]=func()
```

```python
for j in range(N-1):
    n=rn.randint(0,2)
    r[:,j]=(r[:,j-1]+points[n,:])/2
    
    
```

```python
plt.figure(figsize=(7,7))
plt.axes().set_aspect('equal')
plt.plot(r[0,:],r[1,:],'.')
```
